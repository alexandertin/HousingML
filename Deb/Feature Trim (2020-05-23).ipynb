```python
import pandas as pd
import numpy as np

```

```python
iowatrain = pd.read_csv('../data/train.csv', keep_default_na = True)
```

```python
iowatrain.columns.tolist()
```

```python
iowatrain2 = iowatrain.copy()
```

### Remove Columns

BsmtFinSF1 <br>
BsmtFinSF2 <br>
BsmtUnfSF <br>
GarageCars <br>
GarageCond <br>
PoolArea <br>
PoolQC <br>
Fence <br>
MiscFeature <br>

Total: 9 columns

```python
dropcolumns = ['BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','GarageCars','GarageCond','PoolArea','PoolQC','Fence','MiscFeature']
iowatrain2 = iowatrain.drop(columns = dropcolumns)
```

```python
iowatrain2
```

### Remove Rows 

MasVnrType (8 NAs)<br>
MasVnrArea (8 NAs, same as MasVnrType) <br>
Electrical (1 NA)<br>

Total: 9 rows

```python
iowatrain2 = iowatrain2.dropna(subset=['MasVnrType','MasVnrArea','Electrical'])
```

### Mutate

House_Age = 2010 - YearBuilt <br>
Remod_Age = 2010 - YearRemodAdd <br>
Garage_Age = 2010 - GarageYrBlt

```python
iowatrain2 = iowatrain2.assign(House_Age = [max(iowatrain2['YearBuilt'])-x for x in iowatrain2['YearBuilt']])
```

```python
iowatrain2 = iowatrain2.assign(Remod_Age = [max(iowatrain2['YearBuilt'])-x for x in iowatrain2['YearRemodAdd']])
```

```python
iowatrain2 = iowatrain2.assign(Garage_Age = [max(iowatrain2['YearBuilt'])-x for x in iowatrain2['GarageYrBlt']])
```

```python
# Make Garage_Age NAs to 0

iowatrain2['Garage_Age'] = iowatrain2['Garage_Age'].fillna(0)
```

### Factorize
Factorize the following columns

MSSubClass (just factorize) <br>
BsmtQual (No basement as 'None')<br>
BsmtCond (No basement as 'None')<br>
BsmtExposure (No basement as 'None')<br>
BsmtFinType1 (No basement as 'None')<br>
BsmtFinType2 (No basement as 'None')<br>
FireplaceQu (No Fireplace as 'None')<br>
GarageType (No garage as 'None')<br>
GarageFinish (No garage as 'None')<br>
GarageQual (No garage as 'None')<br>

Total: 10 columns

```python
for col in ['MSSubClass','BsmtQual','BsmtCond','BsmtExposure', 'BsmtFinType1','BsmtFinType2','FireplaceQu','GarageType','GarageFinish','GarageQual']:
    iowatrain2[col] = iowatrain2[col].astype('category')
```

```python
iowatrain2['BsmtQual'] = iowatrain2['BsmtQual'].cat.add_categories('None').fillna('None')
iowatrain2['BsmtCond'] = iowatrain2['BsmtCond'].cat.add_categories('None').fillna('None')
iowatrain2['BsmtExposure'] = iowatrain2['BsmtExposure'].cat.add_categories('None').fillna('None')
iowatrain2['BsmtFinType1'] = iowatrain2['BsmtFinType1'].cat.add_categories('None').fillna('None')
iowatrain2['BsmtFinType2'] = iowatrain2['BsmtFinType2'].cat.add_categories('None').fillna('None')
iowatrain2['FireplaceQu'] = iowatrain2['FireplaceQu'].cat.add_categories('None').fillna('None')
iowatrain2['GarageType'] = iowatrain2['GarageType'].cat.add_categories('None').fillna('None')
iowatrain2['GarageFinish'] = iowatrain2['GarageFinish'].cat.add_categories('None').fillna('None')
iowatrain2['GarageQual'] = iowatrain2['GarageQual'].cat.add_categories('None').fillna('None')
```

```python
iowatrain2['GarageQual'].value_counts
```

### Classification

Alley (make binary) (1369/1460 No alley access. Codify as 0.) <br>
Heating (make binary) (1428/1460 are Gas A Type Heating. Codify as 1.) <br>


```python
iowatrain2['Alley'] = iowatrain2['Alley'].fillna(0)
iowatrain2['Alley'] = iowatrain2['Alley'].replace('Grvl',1)
iowatrain2['Alley'] = iowatrain2['Alley'].replace('Pave',1)

iowatrain2['Alley'] = iowatrain2['Alley'].astype('category')
```

```python
iowatrain2['Heating'] = iowatrain2['Heating'].replace({'GasA':1,'GasW':0,'Grav':0,'Wall':0,'OthW':0, 'Floor':0})
```

## Export file

```python
iowatrain2.to_csv('train05-23.csv')
```
