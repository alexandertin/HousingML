```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('ggplot')
from sklearn import preprocessing
import seaborn as sns


pd.set_option('display.max_columns',None, 'display.max_rows',None)
```

```python
iowatrain = pd.read_csv('../data/train.csv', index_col= 'Id')
```

```python
#Count the number of missing values for each feature in the dataset
natable = pd.DataFrame(iowatrain.isna().sum(axis=0).sort_values(ascending=False).rename('Missing values'))

#Provide the total number of rows in the training dataset
natable['Total values']= iowatrain.shape[0]

#Calculate the proportion of missing values to total dataset
natable['Percentage'] = (round(iowatrain.isna().sum()/iowatrain.isna().count(),3)).sort_values(ascending=False)

#Remove any features with no missing values
natable = natable[natable['Missing values'] >0]

#Save the features with missing values into a list
nalist = natable.index.tolist()
```
